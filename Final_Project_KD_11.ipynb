# **Data Angka Konsumsi Ikan**

## **Data Preparation 1**

```python
#menampilkan data
import pandas as pd
df_konsumsi = pd.read_csv("https://raw.githubusercontent.com/lissanuy/Final-Project/main/angkakonsumsiikan.csv")
df_konsumsi
```

## **Report 1**

Dengan menggunakan dataset angka konsumsi ikan diperoleh hasil bahwa angka konsumsi ikan di Indonesia dari tahun 2008 sampai dengan 2012 meningkat setiap tahunnya, namun tingkat konsumsinya terbilang masih rendah. 

```python
#Menampilkan data persediaan ikan untuk konsumsi
data1 = df_konsumsi[df_konsumsi['ParamKonsumsiIkan'] == 'Penyediaan ikan untuk konsumsi per kapita']
data1
```

```python
import matplotlib.pyplot as plt
data1 = df_konsumsi[df_konsumsi['ParamKonsumsiIkan'] == 'Penyediaan ikan untuk konsumsi per kapita']
plt.title('Penyediaan ikan untuk konsumsi per kapita')
plt.bar(data1.Tahun,data1.Nilai, color='plum')
plt.xlabel('Tahun')
plt.ylabel('Kg/Kapita/Tahun')
plt.show()
```

Dari grafik di atas dapat kita ketahui bahwa penyediaan ikan untuk konsumsi selalu meningkat setiap tahunnya. Penyediaan ikan dari tahun 2008, 2009, 2010, 2011, dan 2012 berturut-turut adalah 30.95, 33.51, 38.39, 41.95, dan 46.31(Kg/Kapita/Tahun).

```python
#Menampilkan data konsumsi ikan perkapita
data2 = df_konsumsi[df_konsumsi['ParamKonsumsiIkan'] == 'Konsumsi ikan per kapita']
data2
```

```python
import matplotlib.pyplot as plt
data2 = df_konsumsi[df_konsumsi['ParamKonsumsiIkan'] == 'Konsumsi ikan per kapita']
plt.title('Konsumsi ikan per kapita')
plt.bar(data2.Tahun,data2.Nilai, color='c')
plt.xlabel('Tahun')
plt.ylabel('Kg/Kapita/Tahun')
plt.show()
```

Dari grafik diatas kita ketahui bahwa angka konsumsi ikan perkapita dari tahun 2008 sampai dengan 2012 naik setiap tahunnya. Angka konsumsi ikan terbanyak terjadi pada tahun 2012 dengan nilai konsumsi 33.89 Kg/Kapita/Tahun. 

## **Insight 1**

Untuk memperdalam hasil diatas, akan dianalisis lebih lanjut mengapa tingkat konsumsi ikan di Indonesia masih rendah padahal angka konsumsinya meningkat setiap tahun. 

```python
data=pd.concat([data1,data2],axis=0)
urut=data.sort_index(axis=0)
urut
```

```python
# Plot multi-line chart untuk melihat perbandingan penyediaan ikan dan konsumsi ikan per kapita
import matplotlib.pyplot as plt
urut.groupby(['Tahun','ParamKonsumsiIkan'])['Nilai'].mean().unstack().plot(marker='.', cmap='plasma')
plt.title('Perbandingan Penyediaan dan Konsumsi Ikan Per Tahun',loc='center',pad=30, fontsize=20, color='blue')
plt.xlabel('Tahun', fontsize = 15)
plt.ylabel('Kg/Kapita/Tahun',fontsize = 15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.legend(loc='upper center', bbox_to_anchor=(1.1, 1), shadow=True, ncol=1)
plt.gcf().set_size_inches(12, 5)
plt.tight_layout()
plt.show()
```

Dari grafik di atas, dapat kita ketahui bahwa angka konsumsi ikan dan penyediaan ikan per kapita dari tahun 2008 sampai dengan 2012 sama-sama mengalami peningkatan nilai setiap tahunnya, namun selisih diantara keduanya juga meningkat setiap tahunnya. Maka dari itu, didapatkan kesimpulan bahwa tingkat konsumsi ikan di Indonesia masih rendah karena peningkatan angka konsumsi ikan tidak sebanding dengan peningkatan jumlah penyediaan ikan.

## **Report 4**

Dengan menggunakan dataset produksi olahan ikan, akan dilakukan analisis dengan pertanyaan sebagai berikut:

- 3 Jenis olahan ikan apa yang paling banyak di produksi Indonesia? 
- Berapa rata - rata volume dari ketiga jenis olahan ikan tersebut?

```python
#Mengambil informasi top 3 jenis olahan ikan berdasarkan rata-rata volume 
top_jenis_olahan = (produk_olahan_ikan.groupby('JenisOlahan')['Volume']
                    .mean()
                    .reset_index()
                    .sort_values(by='Volume',ascending=False)
                    .head(3))

#Membuat dataframe baru, filter hanya  hanya top 3 jenis olahan ikan
dataset_top3jenisolahan = produk_olahan_ikan[produk_olahan_ikan['JenisOlahan'].isin(top_jenis_olahan['JenisOlahan'].to_list())]

#Mencetak top 3 jenis olahan ikan
print(top_jenis_olahan)
```

```python
# Membuat Visualisasi Data Top 3 Produk Olahan Ikan di Indonesia
import matplotlib.pyplot as plt #Melakukan import library matplotlib sebagai plt
jenis_olahan = ['Semua Olahan Ikan','Ikan Beku','Ikan Kering/Asin']
volume = [19921.046717, 8102.939394, 7468.832071]

plt.figure(figsize=(12,7))
plt.bar(jenis_olahan, volume, color='lightcoral')

plt.title('Top 3 Produk Olahan Ikan di Indonesia (2001 - 2012)', size=16)
plt.ylabel('Rata-Rata Volume', size=14)
plt.xticks(size=12)
plt.yticks(size=12)

plt.show()
```

Dari grafik di atas dapat diketahui bahwa selama Tahun 2001 - 2012 jenis olahan ikan yang paling banyak di produksi oleh Indonesia adalah Semua Olahan Ikan dengan volume 19921.046717 atau sekitar 20000 m^3, Ikan Beku dengan volume 8102.939394 atau sekitar 8000 m^3, dan Ikan Keirng/Asin dengan volume 7468.832071 atau sekitar 7500 m^3.

## **Insight 6**

Apa yang menyebabkan Perlakuan tersebut dapat menghasilkan Volume Tertinggi dan Mengapa Wilayah tersebut yang menghasilkan Volume Tertinggi?

```python
# Membuat Visualisasi Persebaran Plot Wilayah untuk Mempermudah dalam Mengetahui Perkembangan Produksi Perlakuan Ikan Nasional
import matplotlib.pyplot as plt
from plotnine import *
import plotnine
plotnine.options.figure_size=(10, 4.5)
(ggplot(data=produksi_perlakuan_ikan)
+ aes(y='Volume', x='JenisPerlakuan', color='Wilayah') #Memberi warna
+ geom_point()
+ coord_flip()
+labs(title = 'Perkembangan Produksi Perlakuan Ikan Nasional',
x = 'Perlakuan',
y = 'Volume')
).draw()
plt.show()
```

```python
# Menampilkan Visualisasi Garis untuk Memperkuat Kesimpulan dari Visualisasi Persebaran Plot 
import matplotlib.pyplot as plt
from plotnine import *
import plotnine
plotnine.options.figure_size=(10, 4.5)
(ggplot(data=produksi_perlakuan_ikan)
+ aes(x='JenisPerlakuan', y='Volume', color='Wilayah')
+ geom_line()
+ coord_flip()
+labs(title = 'Perkembangan Produksi Perlakuan Ikan Nasional',
x = 'Perlakuan',
y = 'Volume')
).draw()
plt.show()
```

Berdasarkan Visualisasi yang sudah dilakukan maka diperoleh Kesimpulan bahwa beberapa Jenis Perlakuan sudah digunakan pada Wilayah di Indonesia namun belum menghasilkan Volume Tertinggi, sehingga dengan menggabungkan seluruh Jenis Perlakuan yang ada yaitu dengan menggunakan Jenis Perlakuan Semua diperoleh Volume Tertinggi, dikarenakan Jenis Perlakuan Semua tentunya lebih cocok untuk diterapkan sebab bisa mencoba seluruh Jenis Perlakuan yang lainnya. Jenis Perlakuan Semua pada dasarnya dilakukan untuk meminimalisir ketidaksesuaian Jenis Perlakuan lainnya, karena jika menggunakan satu Jenis Perlakuan misalnya Jenis Perlakuan Peragian tentunya kurang efektif, sebab bisa jadi Jenis Perlakuan Peragian tidak sesuai dengan kondisi Wilayah tersebut. Dapat dilihat pada Visualisasi yang ditampilkan bahwa Wilayah Maluku-Papua menghasilkan Volume Tertinggi, karena Wilayah tersebut telah menggunakan seluruh jenis Perlakuan dengan hasil Volume yang beranekaragam, dapat dilihat pada seluruh Jenis Perlakuan Wilayah tersebut hampir selalu menghasilkan Volume yang cukup tinggi dari tahun ke tahun daripada Wilayah lainnya sehingga ketika seluruh Jenis Perlakuan digabungkan maka akan memberikan hasil Volume yang Tertinggi.
